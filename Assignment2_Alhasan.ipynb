Q8) What is a hash table and how does it relate to dictionaries in python?

Ans: A hash table (or hash map) is a data structure that implements an associative array abstract data type, a structure that can map keys to values.  A hash table uses a hash function to compute an index into an array of buckets or slots, from which the desired value can be found.

In Python, dictionaries are implemented using hash tables.  When you add a key-value pair to a dictionary, Python's hash function calculates the hash of the key. This hash value determines the bucket where the key-value pair will be stored in the underlying hash table.  When you look up a key, Python uses the same hash function to find the correct bucket, and then searches within that bucket for the matching key.

Q17) Describe time complexity of accessing elements in dictionary

Ans:

Accessing elements in a Python dictionary has an average time complexity of **O(1)**, **constant time**.  This is because dictionaries use a hash table implementation.  

Hash tables allow for very fast lookups, insertions, and deletions on average.

Example:

my_dict = {"a": 1, "b": 2, "c": 3}
print(my_dict["b"])  

Output: time O(1)


However, in the worst-case scenario (Linear Time)

(e.g., hash collisions), the time complexity can degrade to **O(n)**, where n is the number of elements in the dictionary.  

This is relatively rare with good hash functions, but it's important to be aware of the possibility.

my_dict = {0: "zero", 2: "two"}

print(my_dict[0])  

Output: In extreme hash collision scenarios, this could take time O(n).




Q18) In what situations list are prefered over dictionary?

Ans: Lists are preferred over dictionaries when:

1. **Order and position matter:**  Lists maintain the order of elements as they are inserted.  Dictionaries, on the other hand, are unordered collections of key-value pairs.  If the order or sequence of items is important to your program's logic, a list is the correct choice.  

2. **Sequential access is primary:** When you need to process elements one after the other, lists provide efficient sequential access using indexing.  Dictionaries are optimized for accessing values based on their keys, not for sequential traversal.  Although you *can* iterate through dictionaries, it's not as efficient as iterating through a list if order matters.

3. **Storing a sequence of homogeneous items:**  Lists are suitable for collections where all elements share the same type or meaning (e.g., a list of numbers, strings, etc.). While you could use a dictionary with numerical keys in this case, it's often unnecessary overhead.


Example:

1)Sensors data like temperature sensor

temperatures = [25, 26, 24, 27, 28]  

Order and position are meaningful

2) Customer Service Queue

queue = ["Customer1", "Customer2", "Customer3"]

print(queue.pop(0))  

Output: Customer1 (First to be served, removed after done)

3) Chronological Timeline

events = ["Project Start", "First Review", "Completion"]

print(events[1])  

Output: First Review






Q19) why are dictionaries considered unordered and how does that affect data retrieval?

Ans:
A dictionary is termed an unordered collection of objects because dictionaries do not maintain any inherent order of the items based on when they were added. In older versions of Python (before 3.7), dictionaries did not preserve insertion order at all. This meant that when you accessed or printed the items, the order could vary, as dictionaries were optimized for fast lookups rather than maintaining order.

Starting from Python 3.7, dictionaries do retain insertion order as an implementation detail, meaning elements appear in the order they were added when you print or iterate over them.

Data retrieval in dictionaries is done using keys.

Accessing values is fast using their associated keys.  Keys must be immutable (numbers, strings, tuples).

my_dict = {"name": "Alhasan", "age": 26, "city": "Nashik"}

print(my_dict["name"])  

Accessing the value associated with the key "name"

Output: Alhasan

print(my_dict["age"])  

Accessing the value associated with the key "age"

Output: 26
